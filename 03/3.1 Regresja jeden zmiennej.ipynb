```python
# Nagłówki

import numpy as np
import matplotlib
import matplotlib.pyplot as pl
import IPython.html.widgets as widgets
from IPython.html.widgets import fixed
from IPython.display import display

import warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore")

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

from IPython.display import display, Math, Latex
```

```python
import csv
reader = csv.reader(open("fires_thefts.txt"), delimiter=",")

x = list()
y = list()
for xi, yi in reader:
    x.append(float(xi))
    y.append(float(yi)) 
    
print("x = ", x[:10])
print("y = ", y[:10])
```

```python
#zad2

def regdots(x, y):    
    fig = pl.figure(figsize=(16*.6,9*.6))
    ax = fig.add_subplot(111)
    ax.scatter(x, y, c='01', s=50, label="Zaleznosc miedzy liczba pozarow a liczba wlaman")
    
    ax.set_xlabel("Liczba pozarow")
    ax.set_ylabel("Liczba wlaman")
    return fig

def legend(fig):
    ax = fig.axes[0]
    handles, labels = ax.get_legend_handles_labels()
    fig.legend(handles, labels, fontsize='15', loc='upper right')
    

fig = regdots(x,y)
legend(fig)
```

```python
#zad3

def h(theta, x):
    return theta[0] + theta[1]*x

def regline(fig, fun, theta, x):
    ax = fig.axes[0]
    x0, x1 = min(x), max(x)
    X = [x0, x1]
    Y = [fun(theta, x) for x in X]
    ax.plot(X, Y, linewidth='3', 
        label=(r'$y=%.2f+%.2f x$' % (theta[0], theta[1])))


fig1 = regdots(x, y)
regline(fig1, h, [10, 1.6], x)
legend(fig1)
```

```python
#zad3 cd

def J(h, theta, x, y):
    m = len(y)
    return (1.0/(2*m) * sum((h(theta, x[i]) - y[i])**2 
                          for i in range(m)))

display(Math(r"\Large J(\theta) = %.4f" % J(h,[10, 1.6],x,y)))
```

```python
#zad 4

def hMx(theta, X):
    return X*theta

def regdotsMx(X, y):    
    fig = pl.figure(figsize=(16*.6,9*.6))
    ax = fig.add_subplot(111)
    ax.scatter(X[:,1], y, c='1', s=80, label="Zaleznosc miedzy liczba pozarow a liczba wlaman")
    
    ax.set_xlabel("Liczba pozarow")
    ax.set_ylabel("Liczba wlaman")
    return fig

def reglineMx(fig, fun, theta, X):
    ax = fig.axes[0]
    x0, x1 = np.min(X[:,1]), np.max(X[:,1])
    L = [x0, x1]
    LX = np.matrix([1, x0, 1, x1]).reshape(2,2)
    ax.plot(L, fun(theta, LX), linewidth='2', 
            label=(r'$y=%.2f+%.2f x$' % (theta[0][0], theta[1][0])));

data = np.loadtxt("fires_thefts.txt", delimiter=",")
m, np1 = data.shape # np1 : n plus 1
n = np1 - 1
Xn = data[:,0:n].reshape(m,n)

XMx = np.matrix(np.concatenate((np.ones((m, 1)), Xn), axis=1)).reshape(m,np1)

yMx = np.matrix(data[:,1]).reshape(m,1)

def JMx(theta,X,y):
    m = len(y)
    J = 1.0/(2.0*m)*((X*theta-y).T*(X*theta-y))
    return J.item()


def LatexMatrix(matrix):
    ltx = r'\left[\begin{array}'
    m, n = matrix.shape
    ltx += '{' + ("r" * n) + '}'
    for i in range(m):
        ltx += r" & ".join([('%.4f' % j.item()) for j in matrix[i]]) + r" \\ "
    ltx += r'\end{array}\right]'
    return ltx

def norm(X,y):
    return (X.T*X)**-1*X.T*y

thetaNorm = norm(XMx, yMx)
display(Math(r'\Large \theta = ' + LatexMatrix(thetaNorm)))
display(Math(r'\Large J(\theta) = %.4f' % JMx(thetaNorm, XMx, yMx)))

fig = regdotsMx(XMx,yMx)
reglineMx(fig, hMx, thetaNorm, XMx)
legend(fig)
```

```python
#zad5

def dJMx(theta,X,y):
    return 1.0/len(y)*(X.T*(X*theta-y)) 

thetaMx2 = np.matrix([-5,1.3]).reshape(2,1)

display(Math(r'\large \theta = ' + LatexMatrix(thetaMx2) + 
             r'\quad' + r'\large \nabla J(\theta) = ' 
             + LatexMatrix(dJMx(thetaMx2,XMx,yMx)))) 

def GDMx(fJ, fdJ, theta, X, y, alpha=0.1, eps=10**-3):
    errorCurr = fJ(theta, X, y)
    errors = [[errorCurr, theta]]
    i = 0
    while True:
        i=i+1
        theta = theta - alpha * fdJ(theta, X, y) # implementacja wzoru
        errorCurr, errorPrev = fJ(theta, X, y), errorCurr
        if abs(errorPrev - errorCurr) <= eps:
            break
        errors.append([errorCurr, theta]) 
    return theta, errors

thetaStartMx = np.matrix([0,0]).reshape(2,1)

thetaBestMx, errors = GDMx(JMx, dJMx, thetaStartMx, 
                       XMx, yMx, alpha=0.0001, eps=0.00001)

display(Math(r'\large\textrm{Wynik:}\quad \theta = ' + 
             LatexMatrix(thetaBestMx) + 
             (r' \quad J(\theta) = %.4f' % errors[-1][0])  
             + r' \quad \textrm{po %d iteracjach}' % len(errors)))  

```

```python
fig = regdotsMx(XMx,yMx)
reglineMx(fig, hMx, thetaBestMx, XMx)
legend(fig)
```

```python
#zad6 wartości alpha zmienione, żeby się to w ogóle wykonało
alphas = [0.00001, 0.0001, 0.001]
```

```python
def GDMx2(fJ, fdJ, theta, X, y, alpha, eps=10**-3):
    errorCurr = fJ(theta, X, y)
    errors = [errorCurr]
    i = 0
    while i < 199:
        i=i+1
        theta = theta - alpha * fdJ(theta, X, y) # implementacja wzoru
        errorCurr, errorPrev = fJ(theta, X, y), errorCurr
        if abs(errorPrev - errorCurr) <= eps:
            break
        errors.append(errorCurr) 
    return errors

fig = pl.figure(figsize=(15,10))
x = np.arange(0, 200, 1)
y1 = GDMx2(JMx, dJMx, thetaStartMx, XMx, yMx, alphas[0], eps=0.00001)
y2 = GDMx2(JMx, dJMx, thetaStartMx, XMx, yMx, alphas[1], eps=0.00001)
y3 = GDMx2(JMx, dJMx, thetaStartMx, XMx, yMx, alphas[2], eps=0.00001)


ax = fig.add_subplot(3,1,1)
line1 = ax.plot(x,y1,color='red',lw=2, label="alpha = " + str(alphas[0]))
line2 = ax.plot(x,y2,color='blue',lw=2, label="alpha = " + str(alphas[1]))
line3 = ax.plot(x,y3,color='green',lw=2, label="alpha = " + str(alphas[2]))
ax.legend()

 
```
