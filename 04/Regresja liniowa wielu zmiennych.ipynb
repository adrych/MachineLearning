```python
#zad2.a

def JMx(theta,X,y):
    m = len(y)
    J = 1.0/(2.0*m)*((X*theta-y).T*(X*theta-y))
    return J.item()


def LatexMatrix(matrix):
    ltx = r'\left[\begin{array}'
    m, n = matrix.shape
    ltx += '{' + ("r" * n) + '}'
    for i in range(m):
        ltx += r" & ".join([('%.4f' % j.item()) for j in matrix[i]]) + r" \\ "
    ltx += r'\end{array}\right]'
    return ltx

def norm(X,y):
    return (X.T*X)**-1*X.T*y

data = np.loadtxt(open("train.csv"), delimiter=",", skiprows=1, usecols=(1,2,3,4))
data[:,[0,3]] = data [:, [3,0]]
data1, data2 = np.split(data, [4000])


m, np1 = data1.shape
n = np1 - 1
Xn = data1[:,0:n].reshape(m,n)

XMx = np.matrix(np.concatenate((np.ones((m, 1)), Xn), axis=1)).reshape(m,np1)

yMx = np.matrix(data1[:,3]).reshape(m,1)

thetaNorm = norm(XMx, yMx)
display(Math(r'\Large \theta = ' + LatexMatrix(thetaNorm)))
display(Math(r'\Large J(\theta) = %.4f' % JMx(thetaNorm, XMx, yMx)))

```

```python
#zad2.b

m, np1 = data2.shape
n = np1 - 1
Xn2 = data2[:,0:n].reshape(m,n)

XMx2 = np.matrix(np.concatenate((np.ones((m, 1)), Xn2), axis=1)).reshape(m,np1)

yMx2 = np.matrix(data2[:,3]).reshape(m,1)

display(Math(r'\Large J(\theta) = %.4f' % JMx(thetaNorm, XMx2, yMx2)))
```

```python
#zad2.c

def rmse(theta, data, y):    
    d = (data*theta)-y
    return sum(d)/len(d)
        
int(rmse(thetaNorm, XMx2, yMx2))
```
