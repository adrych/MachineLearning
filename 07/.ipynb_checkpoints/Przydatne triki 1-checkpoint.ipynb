```python
import numpy as np
import math as math

def runningMeanFast(x, N):
    return np.convolve(x, np.ones((N,))/N, mode='valid')

def safeSigmoid(x, eps=0):
    y = 1.0/(1.0 + np.exp(-x))
    # przytnij od dolu i gory
    if eps > 0:
        y[y < eps] = eps
        y[y > 1 - eps] = 1 - eps
    return y

def h(theta, X, eps=0.0):
    return safeSigmoid(X*theta, eps)

def J(h,theta,X,y):
    m = len(y)
    f = h(theta, X, eps=10**-7)
    return -np.sum(np.multiply(y, np.log(f)) + 
                   np.multiply(1 - y, np.log(1 - f)), axis=0)/m

def dJ(h,theta,X,y):
    return 1.0/len(y)*(X.T*(h(theta,X)-y))

def softmax(X):
    return np.exp(X)/np.sum(np.exp(X))

def softmax(X):
    for x in range(X.shape[0]):
        X[x]=round(X[x],4)
    #print X[0]
    return np.exp(X)/np.sum(np.exp(X))
```

```python
def SGD(h, fJ, fdJ, theta, X, y, 
        alpha=0.0001, maxEpochs=1.0, batchSize=100, adaGrad=True):
    m, n = X.shape
    start, end = 0, batchSize
    eps = 10 ** -7
    hist = np.matrix(np.zeros(n)).reshape(n,1)
    
    maxSteps = (m * float(maxEpochs)) / batchSize
    for i in range(int(maxSteps)):
        XBatch, yBatch =  X[start:end,:], y[start:end].reshape(end-start,1)
 
        fdj = fdJ(h, theta, XBatch, yBatch)
        
        if (adaGrad):
            hist += np.multiply(fdj, fdj)
            adj_grad = np.multiply(alpha / (np.sqrt(hist)+eps), fdj)

        else:
            adj_grad = alpha * fdj
            
        theta = theta - adj_grad
        
        if start + batchSize < m:
            start += batchSize
        else:
            start = 0
        end = min(start + batchSize, m)
        
    return theta
```

```python
import os
import struct
import numpy as np

%matplotlib inline

def read(dataset = "training", path = "."):
    """
    Python function for importing the MNIST data set.  It returns an iterator
    of 2-tuples with the first element being the label and the second element
    being a numpy.uint8 2D array of pixel data for the given image.
    """

    if dataset is "training":
        fname_img = os.path.join(path, 'train-images.idx3-ubyte')
        fname_lbl = os.path.join(path, 'train-labels.idx1-ubyte')
    elif dataset is "testing":
        fname_img = os.path.join(path, 't10k-images.idx3-ubyte')
        fname_lbl = os.path.join(path, 't10k-labels.idx1-ubyte')
    else:
        raise ValueError, "dataset must be 'testing' or 'training'"

    # Load everything in some numpy arrays
    with open(fname_lbl, 'rb') as flbl:
        magic, num = struct.unpack(">II", flbl.read(8))
        lbl = np.fromfile(flbl, dtype=np.int8)

    with open(fname_img, 'rb') as fimg:
        magic, num, rows, cols = struct.unpack(">IIII", fimg.read(16))
        img = np.fromfile(fimg, dtype=np.uint8).reshape(len(lbl), rows, cols)

    get_img = lambda idx: (lbl[idx], img[idx])

    # Create an iterator which returns each image in turn
    for i in xrange(len(lbl)):
        yield get_img(i)
        
def toMatrix(data, maxItems=1000):
    datalist = [t for t in data]
    m = maxItems
    n = 28 * 28 + 1
    X = np.matrix(np.zeros(m * n)).reshape(m, n)
    Y = np.matrix(np.zeros(m)).reshape(m, 1)
    for i, (label, image) in enumerate(datalist[:m]):
        X[i, 0] = 1 # bias term
        X[i, 1:] = image.reshape(28*28,)
        Y[i] = label
    return X, Y
```

```python
data_train = list(read())
data_test = list(read("testing"))

train1000 = toMatrix(data_train, 10000)
test1000= toMatrix(data_test, 1000)

Xtrain = train1000[0][:]
Xtest = test1000[0][:]

mTrain, nTrain = Xtrain.shape
mTest, nTest = Xtest.shape

Xtrain = np.matrix(np.concatenate((np.ones((mTrain, 1)), Xtrain), axis=1)).reshape(mTrain,nTrain+1)
Xtest = np.matrix(np.concatenate((np.ones((mTest, 1)), Xtest), axis=1)).reshape(mTest,nTest+1)

YtrainB = train1000[1][:]
YtestB = test1000[1][:]

YtrainAll = np.zeros((mTrain,10))
YtestAll = np.zeros((mTest,10))



for i in range(0, mTrain):
    for j in range(0,10):
        if YtrainB[i] == j:
            YtrainAll[i,j] = 1
            
for i in range(0, mTest):
    for j in range(0,10):
        if YtestB[i] == j:
            YtestAll[i,j] = 1
        
```

```python
def trainMaxEnt(X, Y):
    n = X.shape[1]
    thetas = []
    for c in range(Y.shape[1]):
        YBi = Y[:,c]
        theta = np.matrix(np.random.random(n)).reshape(n,1)
        #print theta.shape
        thetaBest = SGD(h, J, dJ, theta, X, YBi, 
                       alpha = 1.0, maxEpochs=2.0, batchSize = 50, adaGrad=True)
        thetas.append(thetaBest)
    return thetas

thetas = trainMaxEnt(Xtrain, YtrainAll)
```

```python
def softmax(X):
    X128 = X.astype(np.float128)
    r = np.exp(X128)/np.sum(np.exp(X128))
    return r

def classify(thetas, X):
    regs = np.array([(X*theta).item() for theta in thetas])
    probs = softmax(regs)
    return np.argmax(probs), probs

YTestCls = YtestAll * np.matrix((range(0,10))).T
```

```python
acc = 0.0
for i in range(len(YTestCls)):
    cls, probs = classify(thetas, Xtest[i])
    correct = int(YTestCls[i].item())
    print correct, "<=>", cls, " - ", correct == cls#, probs.tolist()
    
    acc += correct == cls
print "Accuracy =", acc/float(len(Xtest))
```

```python
#prepare data
data_train = list(read())
data_test = list(read("testing"))

train1000 = toMatrix(data_train, 10000)
test1000= toMatrix(data_test, 1000)

Xtrain = train1000[0][:]
Xtest = test1000[0][:]

mTrain, nTrain = Xtrain.shape
mTest, nTest = Xtest.shape

Xtrain = np.matrix(np.concatenate((np.ones((mTrain, 1)), Xtrain), axis=1)).reshape(mTrain,nTrain+1)
Xtest = np.matrix(np.concatenate((np.ones((mTest, 1)), Xtest), axis=1)).reshape(mTest,nTest+1)

YtrainB = train1000[1][:]
YtestB = test1000[1][:]

YtrainAll = np.zeros((mTrain,10))
YtestAll = np.zeros((mTest,10))



for i in range(0, mTrain):
    for j in range(0,10):
        if YtrainB[i] == j:
            YtrainAll[i,j] = 1
            
for i in range(0, mTest):
    for j in range(0,10):
        if YtestB[i] == j:
            YtestAll[i,j] = 1
```

```python
v = 5 #number of validation sets
s = mTrain/v #size of validation set

#randomization

for j in range(1000):
        r1 = np.random.randint(0, s)
        r2 = np.random.randint(0, s)
        temp = Xtrain[r1]
        Xtrain[r1] = Xtrain[r2]
        Xtrain[r2] = temp
        temp = YtrainAll[r1]
        YtrainAll[r1] = YtrainAll[r2]
        YtrainAll[r2] = temp
```

```python
#prepare validation sets

xValidation = np.empty((5, s, 786))
yValidation = np.empty((5, s, 10))

xTrainSet = np.empty((5, mTrain-s, 786))
yTrainSet = np.empty((5, mTrain-s, 10))



for i in range(v):
    xValidation[i] = Xtrain[s*i : s*i+s][:]
    yValidation[i] = YtrainAll[s*i : s*i+s][:]
    
    Xrest1 = Xtrain[: i*s][:]
    Xrest2 = Xtrain[s*i+s:][:]
    
    Yrest1 = YtrainAll[: i*s][:]
    Yrest2 = YtrainAll[s*i+s:][:]
    
    xTrainSet[i] = np.matrix(np.concatenate((Xrest1, Xrest2), axis=0))
    yTrainSet[i] = np.matrix(np.concatenate((Yrest1, Yrest2), axis=0))
    
```

```python
#modifiable training function
def trainMaxEntModified(X, Y, alpha = 1.0, maxEpochs = 2.0, batchSize = 50, adaGrad = True):
    n = X.shape[1]
    thetas = []
    for c in range(Y.shape[1]):
        YBi = Y[:,c]
        theta = np.matrix(np.random.random(n)).reshape(n,1)
        #print theta.shape
        thetaBest = SGD(h, J, dJ, theta, X, YBi, 
                       alpha = 1.0, maxEpochs=2.0, batchSize = 50, adaGrad=True)
        thetas.append(thetaBest)
    return thetas
```

```python
#prepare parameters sets for validation
parameters = np.empty((v, 4))
for i in range(v):
    #alpha, maxEpochs, batchSize, adaGrad
    parameters[i][0] = (1.0*0.1**v%3)
    parameters[i][1] = np.round(v-(1+v/2))
    parameters[i][2] = (50+10*(v))
    parameters[i][3] = v%2
```

```python
#perform cross validation and get the best parameters
accAll = np.empty((v), dtype=np.float64);
bestParams = 0
max = 0
for j in range(v):
    thetas = trainMaxEntModified(xTrainSet[j], yTrainSet[j], parameters[j])
    yValidationCls = yValidation[j] * np.matrix((range(0,10))).T
    acc = 0.0
    for i in range(len(yValidationCls)):
        cls, probs = classify(thetas, xValidation[j][i])
        correct = int(yValidationCls[i].item())
        #print correct, "<=>", cls, " - ", correct == cls, np.round(probs, 4).tolist()

        acc += correct == cls
    print "Accuracy =", acc/float(len(xValidation[j]))
    accAll[j] = acc
    if acc > max:
        bestParams = j
        max = acc
        
parameter = parameters[bestParams]
```

```python
#classify with the best parameters
thetas = trainMaxEntModified(Xtrain, YtrainAll, parameter)
YTestCls = YtestAll * np.matrix((range(0,10))).T

acc = 0.0
for i in range(len(YTestCls)):
    cls, probs = classify(thetas, Xtest[i])
    correct = int(YTestCls[i].item())
    print correct, "<=>", cls, " - ", correct == cls#, probs.tolist()
    
    acc += correct == cls
print "Accuracy =", acc/float(len(Xtest))
```
